```python
import requests
from bs4 import BeautifulSoup
import pymongo
from datetime import datetime
import time
```

```python
count = 0

while True:
    print('Initiating loop: '+ str(count))
    # Requesting the site
    r = requests.get("https://merolagani.com/LatestMarket.aspx")
    # Making the soup of site
    soup = BeautifulSoup(r.text, 'html.parser')

    # Extracting table 
    result = soup.find(attrs={'class': 'table table-hover live-trading sortable'})

    # Removing the table head
    result.thead.decompose()

    # Replacing in between newline texts with \n
    whole = result.get_text(separator="\n", strip=True)

    # Splitting the whole 
    whole_list = whole.split('\n')


    # Making list of lists for each companies
    individual_list = []
    for i in whole_list:
        temp = whole_list[:8]

        # converting string numerical values into float type
        for j in range(1,8):
            try:
                #print(temp[i])
                temp[j] = float(temp[j].replace(',', ''))
                #print(temp[i])
            except:
                break
        if (len(temp) == 0):
            break
        #print(temp)
        whole_list = whole_list[8:]
        individual_list.append(temp)


    # Establishing connection with the mongodb
    mongodb_client = pymongo.MongoClient("mongodb+srv://susant:susant@cluster0.yevtz.mongodb.net/Merolagani?retryWrites=true&w=majority")

    merolaganiDb = mongodb_client['Merolagani']
    collection = merolaganiDb["test"]

    # Inserting the data in the Mongodb
    for each in individual_list:
        #print(each[0])
        data = {"DateTime": datetime.now(),"Symbol": each[0],"LTP":each[1], "LTV": each[2], "% Change": each[3], "High": each[4], "Low": each[5], "Open": each[6], "Qty": each[7]}
        collection.insert_one(data)
        
    count = count + 1
    
    # Breaking out of loop
    if count > 14:
        print('Breaking Loop: Successful!!!')
        break
    
    # Sleeping for Five Minutes
    print('Sleeping for Five Minutes')
    time.sleep(300)
    
    
```
